```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dt = 0.001
T = 1.
n = int(T/dt)
t = np.linspace(0.,T,n)
x = np.zeros(n)
sigma = 0.5
tau = 0.05
mu = 10
ttau = np.sqrt(2./tau)
sdt = np.sqrt(dt)

for i in range(1,n):
    x[i] = x[i-1] - ((x[i-1] - mu)/tau)*dt + sdt*ttau*np.random.randn()*sigma

fig, ax = plt.subplots(1,1,figsize=(8,4))    
ax.plot(t,x,lw=2)
```
